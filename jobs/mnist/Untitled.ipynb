```python
def get_param_from_string(param_str):
    match_digit = "\((\d)\)"
    match_int = "\((\d+)\)"
    match_str = "\((\w+)\)"
    match_float = "\((\d+.\d+)\)"
    regex = "mnist_batchnorm"+match_digit+"_lr"+match_float+"_momentum"+match_float+"_wdecay"+match_float+"_n_clusters"+match_int+"_n_cycles"+match_int+"_rnd"+match_int+"_t_batch_size"+match_int+"_shuffle"+match_digit+"_pca"+match_str+"_"
    m = re.match(regex, param_str)
    params = {}
    params["batchnorm"] = bool(m.group(1))
    params["lr"] = float(m.group(2))
    params["momentum"] = float(m.group(3))
    params["wdecay"] = float(m.group(4))
    params["n_clusters"] = int(m.group(5))
    params["n_cycles"] = int(m.group(6))
    params["random state"] = int(m.group(7))
    params["t_batch_size"] = int(m.group(8))
    params["shuffle"] = bool(m.group(9))
    params["pca"] = int(m.group(10)) if not m.group(10) == "None" else "None"
    return params

def get_validation_from_tensorboard(tensorboard_file):
    
```
